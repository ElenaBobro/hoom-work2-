```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/file csv/yulu_rental.csv')
```

1. Як змінюється загальна кількість орендованих велосипедів протягом року в залежності від сезону?

```python
seasonal_counts = df.groupby('season')['count'].mean()

seasonal_counts.plot(kind='bar')
plt.title('Середня кількість орендованих велосипедів по сезонам')
plt.xlabel('Сезон')
plt.ylabel('Середня кількість орендованих велосипедів')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Весна', 'Літо', 'Осінь', 'Зима'], rotation=0)
plt.show()
```

На діаграмі можна побачити, що весною число орендованих велосипедів найменше. Для збільшення попиту на оренду можна заповадити сезонні акції або якусь бонусну систему.

 2. Як температура впливає на кількість орендованих велосипедів?

```python
plt.scatter(df['temp'], df['count'], alpha=0.5)
plt.title('Кореляція між температурою та кількістю орендованих велосипедів')
plt.xlabel('Температура (°C)')
plt.ylabel('Кількість орендованих велосипедів')
plt.show()
```

Найбільша кількість орендованих велосипедів фіксується при комфортних умовах (температура від 15 до 30°C), що може вказувати на залежність оренди від більш сприятливих погодних умов.

Як різні погодні умови впливають на кількість орендованих велосипедів?


```python
weather_counts = df.groupby('weather')['count'].mean()

weather_counts.plot(kind='bar')
plt.title('Середня кількість орендованих велосипедів за погодними умовами')
plt.xlabel('Погодні умови')
plt.ylabel('Середня кількість орендованих велосипедів')
plt.xticks(rotation=45)
plt.show()
```

Погодні умови відіграють важливу роль на кількість орендованих велосипедів. Найвищі показники оренди при гарних погодних умовах. Ця діаграма може допомогти у визначенні періодів для проведення акцій або підвищення доступності велосипедів у несприятливу погоду.
